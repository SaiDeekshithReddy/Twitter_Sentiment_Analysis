```
import numpy as np
import pandas as pd
```

```
data = pd.read_csv('train_tweets.csv', index_col=0)
```

```
data
```

```
stri = "we won!!! love the land!!! #allin #cavs #champions #cleveland #clevelandcavaliers  â¦"
stri
```

```
import re
lst_wrds = re.sub("(@[A-Za-z0-9]+)|([^A-Za-z0-9 \t])", " ", stri).split()
lst_wrds
```

```
" ".join(lst_wrds)
```

```
def processed_tweet(tweet):
  lst_wrd = re.sub("(@[A-Za-z0-9]+)|([^A-Za-z0-9 \t])", " ", tweet).split()
  processed_tweet = " ".join(lst_wrd)
  return processed_tweet
```

```
data['process_tweet'] = data['tweet'].apply(processed_tweet)
```

```
data.head()
```

```
from sklearn.model_selection import train_test_split
```

```
x = data['process_tweet']
y = data['label']
```

```
x
```

```
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```
x_train.shape
```

```
x_test.shape
```

```
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
```

```
cnt_vect = CountVectorizer(stop_words='english')
```

```
x_train_cnts = cnt_vect.fit_transform(x_train)
```

```
x_train_cnts
```

```
tdifd = TfidfTransformer(norm='l2')
```

```
x_tfidf = tdifd.fit_transform(x_train_cnts)
```

```
x_tfidf
```

```
y_train.shape
```

logistic

```
from sklearn.linear_model import LogisticRegression
```

```
model = LogisticRegression()
model.fit(x_tfidf, y_train)
```

```
x_test_cnts = cnt_vect.transform(x_test)
x_test_cnts
```

```
x_test_tf = tdifd.fit_transform(x_test_cnts)
```

```
pred = model.predict(x_test_tf)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```
